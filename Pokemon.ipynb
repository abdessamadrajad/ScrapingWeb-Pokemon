# Web Scraping with Requests and Beautiful Soup "BS4" #Pokemon Exercice

### Import needed libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://pokemondb.net/pokedex/all'
```

### Requests

```python
req = requests.get(url)
status = req.status_code
encoding = req.encoding
html_doc = req.text
status
```

### BS4

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
names=soup.find_all("a", {"class":"ent-name"})
print(len(names))
type(names)
```

```python
list_nom= []
for item in names:
    list_nom.append(item.getText())
    
len(list_nom)
```

```python
#removing redundants items
```

```python
list_nom=list(dict.fromkeys(list_nom))
```

```python
len(list_nom)
```

```python
Id=soup.find_all("span", {"class":"infocard-cell-data"})
len(Id)
```

```python
list_id= []
for i in Id:
    list_id.append(i.getText())
list_id=list(dict.fromkeys(list_id))
len(list_id)


```

```python
### Save data in csv 
```

```python
import pandas as pd
dataframe=pd.DataFrame(columns = ['id', 'Name']) 
dataframe.id=list_id
dataframe.Name=list_nom
##dataframe.to_csv (r'pokemon_dataframe.csv', index = False, header=True)

dataframe
```

```python
image=soup.find_all("span", {"class":"infocard-cell-img"})
```

```python
len(image)
```

```python
list_image= []
for itm in image:
    list_image.append(itm.span.get("data-src"))
    
list_image
```

```python
def get_img(url_img,id):
    r = requests.get(url_img)
    if r.status_code == 200:
        img_container="image_"+str(id)+".jpg"
        with open(img_container, 'wb') as f:
            f.write(r.content)
```

```python
index=0
for url in list_image:
    get_img(url,index)
    index+=1
```
